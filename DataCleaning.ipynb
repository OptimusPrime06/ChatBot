## Data Cleaning

### Riddles dataset cleaning

```python
cleanRdf= rdf.drop(columns='Hint')
cleanRdf
```

### Clean Conversation dataset

```python
cleanCdf = cdf.drop(columns=[col for col in cdf.columns if col.startswith('Unnamed')])
cleanCdf 
```

### Adding riddles to our custom dataset

```python

# Iterate over the range of indices
for i, puzzle in enumerate(cleanRdf['Riddle']):
    # Calculate the target index in customDf
    target_index = customDf.index.max()+1
    # Assign the riddle to the Statement column in customDf
    customDf.loc[target_index, 'Statement'] = puzzle
    customDf.loc[target_index, 'Intent'] = 'puzzle'
    customDf.loc[target_index, 'DrowsinessLevel'] = 'low'
    customDf.loc[target_index, 'RepFlag'] = 0



for i, answers in enumerate(cleanRdf['Answer']):
    # Calculate the target index in customDf
    target_index = customDf.index.max()+1
    # Assign the riddle to the Statement column in customDf
    customDf.loc[target_index, 'Statement'] = answers
    customDf.loc[target_index, 'DrowsinessLevel'] = 'low'
    customDf.loc[target_index, 'Intent'] = 'answer'
    customDf.loc[target_index, 'RepFlag'] = 0



print(customDf.loc[81:100])
```

### Adding conversations to our custom dataset

```python
# Iterate over the range of indices
for i, questions in enumerate(cleanCdf['question']):
    # Calculate the target index in customDf
    target_index = customDf.index.max()+1
    # Assign the riddle to the Statement column in customDf
    customDf.loc[target_index, 'Statement'] = questions
    if questions.strip().endswith('?'):
        customDf.loc[target_index, 'Intent'] = 'question'
    else:
        customDf.loc[target_index, 'Intent'] = 'answer'
    customDf.loc[target_index, 'DrowsinessLevel'] = 'medium'
    customDf.loc[target_index, 'Intent'] = 'answer'
    customDf.loc[target_index, 'RepFlag'] = 0

for i, answers in enumerate(cleanCdf['answer']):
    # Calculate the target index in customDf
    target_index = customDf.index.max()+1
    # Assign the riddle to the Statement column in customDf
    customDf.loc[target_index, 'Statement'] = answers
    if answers.strip().endswith('?'):
        customDf.loc[target_index, 'Intent'] = 'question'
    else:
        customDf.loc[target_index, 'Intent'] = 'answer'
    customDf.loc[target_index, 'DrowsinessLevel'] = 'medium'
    customDf.loc[target_index, 'Intent'] = 'answer'
    customDf.loc[target_index, 'RepFlag'] = 0

print(customDf.loc[9100:9110])
```
